````python
# Assuming you have imgaug
%matplotlib inline
from util import display_results
````

````python
import cv2
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

def get_faces(image_file):
    img = cv2.imread(image_file)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    return faces
````

## Pulling image from URL

````python
import requests
import shutil
````

````python
sample_image_url = 'https://mountainviewlife.org/wp-content/uploads/2012/02/group.jpg'
````

````python
def download_image(url, output_file = 'temp.jpg'):
    response = requests.get(url, stream=True)
    with open(output_file, 'wb') as out_file:
        shutil.copyfileobj(response.raw, out_file)
````

````python
download_image(sample_image_url)
````

````python
faces = get_faces('temp.jpg')
display_results('temp.jpg', faces)
````
